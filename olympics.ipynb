```python
def fetch_medall_tally(year, country):
    # Initialize temp_df as None
    temp_df = None
    flag =0
    
    # Both year and country are "overall"
    if year == "overall" and country == "overall":
        temp_df = medal_df
    
    # Year is "overall" but a specific country is given
    elif year == "overall" and country != "overall":
        flag =1
        temp_df = medal_df[medal_df["region"] == country]
    
    # A specific year is given but country is "overall"
    elif year != "overall" and country == "overall":
        temp_df = medal_df[medal_df["Year"] == year]
    
    # Both year and country are specific
    elif year != "overall" and country != "overall":
        temp_df = medal_df[(medal_df["Year"] == year) & (medal_df["region"] == country)] 
    # Grouping and sorting
    if flag == 1:
        result = temp_df.groupby("Year")[["Gold", "Bronze", "Silver"]].sum().sort_values(by="Year").reset_index()
    else:
       result = temp_df.groupby("region")[["Gold", "Bronze", "Silver"]].sum().sort_values(by="Gold", ascending=False).reset_index()
    return result  
```

```python
fetch_medall_tally(year=2016, country="overall")
```

```python
def country_year_list (df):
    years = df["Year"].unique().tolist()
    years.sort()
    years.insert(0,"overall")
    country = df["region"].dropna().unique().tolist()
    country.sort()
    country.insert(0,"overall")
    return years, country
```
